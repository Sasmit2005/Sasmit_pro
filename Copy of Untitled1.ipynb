```python

import matplotlib.pyplot as plt

# Select only numeric columns
numeric_df = df.select_dtypes(include='number')

fig, axs = plt.subplots(nrows=len(numeric_df.columns), figsize=(10, 5 * len(numeric_df.columns)))

# Handle the case where there's only one subplot (axs is not iterable)
if len(numeric_df.columns) == 1:
    axs = [axs]

for i, col in enumerate(numeric_df.columns):
    axs[i].boxplot(numeric_df[col], vert=False)
    axs[i].set_ylabel(col)

plt.tight_layout()
plt.show()
```

```python


# Set style
sns.set_style("whitegrid")

# Create the FacetGrid with your existing DataFrame
g = sns.FacetGrid(df, hue="Species", height=6)

# Map the scatter plot
g.map(plt.scatter, 'SepalLengthCm', 'PetalLengthCm').add_legend()

# Show the plot
plt.show()
```
